## Finding polynomial coefficients

Here, I get the coefficients of all 5 polynomials, as I want to see which one gives me the best fit.

```python
fit_2 = np.polyfit(time,centerline_v,2) # Polynomial of order 2
fit_3 = np.polyfit(time,centerline_v,3) # Polynomial of order 3
fit_4 = np.polyfit(time,centerline_v,4) # Polynomial of order 4
fit_5 = np.polyfit(time,centerline_v,5) # Polynomial of order 5
```

## Building the fitted data for every polynomial

```python
trend_2 = []
trend_3 = []
trend_4 = []
trend_5 = []

for i in time:

    trend_2.append(fit_2[0]*i**2 + fit_2[1]*i + fit_2[2])
    trend_3.append(fit_3[0]*i**3 + fit_3[1]*i**2 + fit_3[2]*i + fit_3[3])
    trend_4.append(fit_4[0]*i**4 + fit_4[1]*i**3 + fit_4[2]*i**2 + fit_4[3]*i + fit_4[4])
    trend_5.append(fit_5[0]*i**5 + fit_5[1]*i**4 + fit_5[2]*i**3 + fit_5[3]*i**2 + fit_5[4]*i + fit_5[5])
```

## At what $Re_{tr}$ does the flow transition from the laminar state?

After examining the figures above, flow is estimated to transition at around a reynolds number of 360.

$$ Re_{tr} = 360$$

We can also find an estimate for the standard deviation of the laminar flow $\sigma_{laminar}$ by averaging the running standard deviation before a reynolds number of 360, as seen in the code block below.

For this example, I will use the running standard deviation with window of 600

```python
std_laminar = [ ]

for i in range(len(re)):
    
    if re[i] < 300:
        
        std_laminar.append(std_window_600[i]) # Getting all standard deviation values below 300
        
    else:
        
        continue
        
std_laminar_avg = stts.mean(std_laminar) # Averaging the values
```

And so, the standard deviation of laminar flow has an estimate value of

$$\sigma_{laminar} = 0.0005926076971859508 $$

## Fit a trend to the data in the form

$$ \sigma - \sigma_{laminar} = A (Re-Re_{tr})^x $$

## My solution
To simplify things a bit, I will assign a new variable $\alpha$ such that

$$ \alpha = \sigma - \sigma_{laminar}. $$

I will also assign a new variable $\beta$ such that

$$ \beta = Re - Re_{tr}. $$
Now, I have to fit a powerlaw into this relationship

$$ \alpha = A\beta^x, $$

where $A$ and $x$ are constants.

To simplify things a bit, I can convert this power relationship into a linear one by getting the logarithm of right and left hand side of the above relationship. This yields

$$ log(\alpha) = log(A) + xlog(\beta)$$

This simplifies things a lot, because now the above relationship is a linear one, and all I have to do is find the intercept $log(A)$, and the slope $x$.

Here is a plot of the data which I want to fit a powerlaw fit into.


```python
plt.figure()
plt.xlabel("Reynolds Number")
plt.ylabel("Standard Deviation")
comp_re_l = [ ]
com_std = [ ]
re_l = re.tolist()
for i in range(len(re_l)):
    if re_l[i] > 360 and re_l[i] < 380:
       comp_re_l.append(re_l[i]) 
       com_std.append(std_window_600[i])     
    else:
        continue
        
sns.lineplot(x=comp_re_l,y=com_std)
```

The process of finding that log relationship mentioned earlier is demonstrated in the code block below

```python
alpha = np.array(com_std) - std_laminar_avg
beta = np.array(comp_re_l) - 360

linear_fit = np.polyfit(np.log(abs(beta)),np.log(alpha),1) # Linear fit between log(beta) and log(alpha)

# Now we can plot the linear relationship between alpha and beta

linear_fit_data = [ ]

for i in range(len(alpha)):
    
    linear_fit_data.append(linear_fit[0] * np.log(beta[i]) + linear_fit[1])
    
    
plt.xlabel("log(beta)")
plt.ylabel("log(alpha)")
sns.lineplot(x = np.log(beta),y=np.log(alpha),label="Data")
sns.lineplot(x=beta,y=linear_fit_data,color='tomato',label="Trend")
```
